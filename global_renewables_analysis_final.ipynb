# Global Solar & Wind Energy Trends (2016 - 2020)

The global energy transition from historical use of fossil fuel based electrical generation to renewable generation has increased in pace over the past decade with the advancement in technology, access to capital, and climate change affecting all life in real time. As a potential investor in future solar and wind generation projects we want to understand/analyze global trends in electrical production by solar energy, wind energy, and the investment in these generation capacities via answering the following questions:

    1. During the years 2016 and 2020 what countries produced the most electricity by solar and wind technologies?
    
    2. During the years 2016 and 2020 what countries had the most investment in renewable energy overall?
    
    3. During these same years does the highest amount of solar and wind electricity correspond to the highest investment in renewables?
    
    4. With the results of the above three questions, where might renewable energy investment be a good idea in the near future if we wanted to see the most solar/wind electrical return for our investment?

# Visualization & Analysis

```python
# set up dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Solar & Wind

```python
# read in combined_solar_wind csv as df
solar_wind_df = pd.read_csv("resources/combined_solar_wind.csv", encoding='utf-8')
# change Year column to dtype str
solar_wind_df = solar_wind_df.astype({"Year": "str"}, errors='raise')
solar_wind_df.head()
```

```python
# create an only 2016 df to then sort and visualize to find top producers
only_sixteen = solar_wind_df[solar_wind_df['Year'] == '2016']
only_sixteen.head()
```

```python
# Top 15 solar producers in 2016

# sorting 2016 df by solar Quantity
high_to_low_solar = only_sixteen.sort_values("Solar Quantity (kWh, Million)", ascending=False)
# reset index for use with iloc
high_to_low_solar.reset_index(drop=True, inplace=True)
# use iloc to get top 15 only
high_to_low_solar = high_to_low_solar.iloc[:15]
high_to_low_solar
```

```python
# top 15 wind producers in 2016

# sorting 2016 df by Wind Quantity
high_to_low_wind = only_sixteen.sort_values("Wind Quantity (kWh, Million)", ascending=False)
# reset index for use with iloc
high_to_low_wind.reset_index(drop=True, inplace=True)
# use iloc to get top 15 only
high_to_low_wind = high_to_low_wind.iloc[:15]
high_to_low_wind
```

```python
#bar chart 2016 Solar 
# used a horizontal bar chart for better readabiliity w/methods
# ax.barh and ax.invert_yaxis

y_axis = np.arange(len(high_to_low_solar))
tick_locations = [value for value in y_axis]

fig0, ax0 = plt.subplots()
plt.figure(figsize=(15,6))
ax0.barh(y_axis, high_to_low_solar["Solar Quantity (kWh, Million)"], color='tab:orange',
        alpha=0.8, align="center")
ax0.set_yticks(tick_locations, high_to_low_solar["Country or Area"], rotation="horizontal")
ax0.set_title("Fig 1. Top 15 Countries by Solar Production (2016)")
ax0.set_ylabel("Country or Area")
ax0.set_xlabel("Solar Quantity (kWh, Million)")
ax0.invert_yaxis()
fig0.tight_layout()
fig0.savefig("resources/top15_solar_2016.png")
plt.show()
```

```python
#bar chart 2016 wind 
# used a horizontal bar chart for better readabiliity w/methods
# ax.barh and ax.invert_yaxis

y_axis = np.arange(len(high_to_low_wind))
tick_locations = [value for value in y_axis]

fig1, ax1 = plt.subplots()
plt.figure(figsize=(15,6))
ax1.barh(y_axis, high_to_low_wind["Wind Quantity (kWh, Million)"], color='tab:cyan',
        alpha=0.8, align="center")
ax1.set_yticks(tick_locations, high_to_low_wind["Country or Area"],rotation="horizontal")
ax1.set_title("Fig 2. Top 15 Countries by Wind Production (2016)")
ax1.set_ylabel("Country or Area")
ax1.set_xlabel("Wind Quantity (kWh, Million)")
ax1.invert_yaxis()
fig1.tight_layout()
fig1.savefig("resources/top15_wind_2016.png")
plt.show()
```

## Investment

```python
# read in investment_totals csv as df
investment_total_df = pd.read_csv("resources/investment_totals.csv", encoding='utf-8')
# change Year column to dtype str
investment_total_df = investment_total_df.astype({"Year": "str"}, errors='raise')
investment_total_df.head()
```

```python
# 2016 investment values
investment_sixteen = investment_total_df[investment_total_df['Year'] == '2016']
investment_sixteen.head()
```

```python
#top 15 investors in the world in 2016

# sorting 2016 df by investment
sixteen_investment = investment_sixteen.sort_values("Investment (USD Million)", ascending=False)
# reset index for use with iloc
sixteen_investment.reset_index(drop=True, inplace=True)
# use iloc to get top 15 only
sixteen_investment = sixteen_investment.iloc[:15]
sixteen_investment
```

```python
#only 2016 bar chart investment


y_axis = np.arange(len(sixteen_investment))
tick_locations = [value for value in y_axis]

fig4, ax4 = plt.subplots()
plt.figure(figsize=(6,15))
ax4.barh(y_axis, sixteen_investment["Investment (USD Million)"], color='tab:green',
        alpha=0.8, align="center")
ax4.set_yticks(tick_locations, sixteen_investment["Country or Area"],rotation="horizontal")
ax4.set_title("Fig 5. Top 15 Countries/Areas\nRenewable Energy Investment (2016)")
ax4.set_ylabel("Country or Area")
ax4.set_xlabel("Investment (USD Millions)")
ax4.invert_yaxis()
fig4.tight_layout()
fig4.savefig("resources/top15_invest_2016.png")
plt.show()
```

```python
#only 2020 bar chart investment 


y_axis = np.arange(len(twenty_investment))
tick_locations = [value for value in y_axis]

fig5, ax5 = plt.subplots()
plt.figure(figsize=(15,6))
ax5.barh(y_axis, twenty_investment["Investment (USD Million)"], color='tab:green', 
        alpha=0.8, align="center")
ax5.set_yticks(tick_locations, twenty_investment["Country or Area"],rotation="horizontal")
ax5.set_title("Fig 6. Top 15 Countries/Areas\nRenewable Energy Investment (2020)")
ax5.set_ylabel("Country or Area")
ax5.set_xlabel("Investment (USD Millions)")
ax5.invert_yaxis()
fig5.tight_layout()
fig5.savefig("resources/top15_invest_2020.png")
plt.show()
```

### 2020 Investment Summary & Analysis

Fig 6.( Top 15 Countries of Renewable Energy Investment 2020)

In 2020, Brazil became the largest investor in Renewable Energy in the World . Currently Brazil capacity to generate renewable energy is approximately 84%, higher than the world average of 38%. ## Comparison Analysis

```python
# inner merge both investment df's to see which countries have data for both years

invest_compare = twenty_investment.merge(sixteen_investment, how="inner", on="Country or Area")
invest_compare
```

```python
# inner merge 2020 solar df (has solar/wind top 15 in it already) and 2020 investment df to see which countries
# were in the top 15 solar/wind producers as well as renewable investment recipients

invest_renew_20 = twenty_investment.merge(twenty_hi_lo_solar, how="inner", on="Country or Area")
invest_renew_20 = invest_renew_20.drop(columns=["Year_x","Year_y"])
invest_renew_20
```

```python
# change column names to not only represent metric but year so they can be merged with 2016 later

invest_renew_20.rename(columns = {'Investment (USD Million)': '2020 Investment (USD Million)',
                                 'Solar Quantity (kWh, Million)': '2020 Solar Quantity (kWh, Million)',
                                 'Wind Quantity (kWh, Million)': '2020 Wind Quantity (kWh, Million)'}, inplace = True)
invest_renew_20
```

```python
# inner merge 2016 wind df (has solar/wind top 15 in it already) and 2016 investment df to see which countries
# were in the top 15 solar/wind producers as well as renewable investment recipients

invest_renew_16 = sixteen_investment.merge(high_to_low_wind, how="inner", on="Country or Area")
invest_renew_16 = invest_renew_16.drop(columns=["Year_x","Year_y"])
invest_renew_16
```

```python
# change column names to not only represent metric but year so they can be merged with 2020 later

invest_renew_16.rename(columns = {'Investment (USD Million)': '2016 Investment (USD Million)',
                                 'Solar Quantity (kWh, Million)': '2016 Solar Quantity (kWh, Million)',
                                 'Wind Quantity (kWh, Million)': '2016 Wind Quantity (kWh, Million)'}, inplace = True)
invest_renew_16
```

```python
# inner merge 2016 solar/wind/investment df with 2020 solar/wind/investment df
# to find those countries that were top producers in solar and wind in 2016 and 2020
# as well as top recipients of renewable investment in those same years

invest_renew_total = invest_renew_16.merge(invest_renew_20, how="inner", on="Country or Area")
invest_renew_total
```

### Comparison Summary & Analysis

From the data we compiled there were only 3 countries who were not only top producers of wind and solar in 2016, 2020, but also top recipients of renewable investment; India, Brazil, and Türkiye. These countries should be the focus of possible future investment. 

All three of these countries are rapidly developing and some of the most influential societies in their respective global regions. India's population is the only other country with China that is over 1 Billion so their steep increase, especially in solar, makes a lot of sense.

Fig 7. and Fig 8. Solar vs Wind production of India, Brazil, Türkiye in 2016, 2020

Wind power technology harnesses about 50% of the energy that passes through them while solar power technology harnesses only about 20%. With this is in mind, it is interesting to see that in 2016 wind power was most popular for all three countries, but by 2020 solar was starting to catch up, especially in India. It would be a good idea to dig into the possible why's of this phenomena for knowledge about future investment potential.

Fig 9. Renewable Energy Investment change in those same countries 2016 - 2020

This last chart is possibly the most interesting. While solar and wind electrical production is increasing over the time period 2016 to 2020 for the three above mentioned countries, investment is either relatively stagnant or declining during the same time horizon. One possible explanation is that production lags behind investment. One invests money in a technology in 2016 and then doesn't see the results until 2020. Regardless, this system/situation seems to be more complex than this initial data analysis is able to insightfully show and means more data and more analysis is needed in order to make a truly informed investment decision.
